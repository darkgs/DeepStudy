```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def show_images(images_to_show):
    plt.figure(figsize=(8, 8))
    grid = gridspec.GridSpec(len(images_to_show), len(images_to_show[0]),
                             top=1., bottom=0., right=1., left=0.,
                             wspace=0.0, hspace=0.0)
    for i, image_l in enumerate(images_to_show):
        for j, image in enumerate(image_l):
            ax = plt.subplot(grid[i,j])
            ax.imshow(image)
            ax.set_xticks([])
            ax.set_yticks([])
            ax.set_aspect('auto')
```

```python
from mnist_gan import MNIST_GAN

params = {
    'batch_size': 64,
    'entropy_dim': 100,
    'hidden_dim': 128,
    'lr': 1e-3,
}

mnist_gan = MNIST_GAN(params)
mnist_gan.load(epoch=200)

img_to_show = []
for _ in range(10):
    img_row = []
    for _ in range(10):
        img_row.append(mnist_gan.test())
        
    img_to_show.append(img_row)

show_images(img_to_show)
```
